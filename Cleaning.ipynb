# Cleaning of dataset and aggregating numbers

- Community District spatial data found [here](https://www.nyc.gov/content/planning/pages/resources/datasets/community-districts)
- NYC facilities data found [here](https://data.cityofnewyork.us/City-Government/Facilities-Database/ji82-xba5/about_data)

## Importing packages and loading data

```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('Data/Facilities_Database_20250917.csv')
df
```

```python
df.info()
```

```python
df1 = df[['facname','city', 'optype', 'facgroup', 'latitude', 'longitude']]
df1 = df1[df1['optype'] == 'Public']
df1
```

```python
df1.info()
```

```python
gdf = gpd.read_file('Data/nycd_25c/nycd.shp')
gdf
```

```python
names = pd.read_csv('Data/District_Names.csv')
names.drop(columns=['BOROUGH_ID', 'BOROUGH_ID.1'], inplace=True)
```

```python
gdf_named = pd.merge(gdf, names, right_on="DISTRICTCODE", left_on="BoroCD", how="left")
gdf_named
```

```python
gdf_named = gdf_named[gdf_named['CD Name'].isna() == False]
gdf_named.info()
```

```python
print(gdf_named.crs)
```

## Merging datasets

```python
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(df1["longitude"], df1["latitude"])]
gdf_points = gpd.GeoDataFrame(df1, geometry=geometry, crs="EPSG:4326").to_crs("EPSG:2263")
gdf_points
```

```python
gdf_merged = gpd.sjoin(gdf_points, gdf_named, how="inner", predicate="within")
gdf_merged
```

```python
gdf_merged.info()
```

```python
gdf_merged.drop(columns = ['DISTRICTCODE', 'BOROUGH', 'optype', 'city', 'index_right'])
```

```python
gdf_merged['facgroup'].value_counts()
```

```python
gdf_merged["facgroup"] = gdf_merged["facgroup"].replace({
    "ADULT SERVICES" : "HEALTH AND HUMAN SERVICES",
    "HIGHER EDUCATION" : "EDUCATION",
    "SCHOOLS (K-12)" : "EDUCATION",
    "MATERIAL SUPPLIES AND MARKETS" : "CITY OPERATIONS",
    "OTHER PROPERTY" : "OTHER",
    "TELECOMMUNICATIONS" : "CITY OPERATIONS",
    "HISTORICAL SITES" : "PARKS AND RECREATION",
    "JUSTICE AND CORRECTIONS" : "PUBLIC SAFETY",
    "HUMAN SERVICES" : "HEALTH AND HUMAN SERVICES",
    "HEALTH CARE" : "HEALTH AND HUMAN SERVICES",
    "PARKS AND PLAZAS" : "PARKS AND RECREATION",
    "CITY AGENCY PARKING, MAINTENANCE, AND STORAGE" : "CITY OPERATIONS",
    "OFFICES, TRAINING, AND TESTING" : "CITY OPERATIONS",
    "DAY CARE AND PRE-KINDERGARTEN" : "EDUCATION",
    "EMERGENCY SERVICES" : "PUBLIC SAFETY",
    "LIBRARIES" : "EDUCATION",
    "SOLID WASTE" : "CITY OPERATIONS",
    "WATER AND WASTEWATER" : "CITY OPERATIONS"
})
```

```python
gdf_merged = pd.get_dummies(gdf_merged, columns=["facgroup"])
bool_cols = gdf_merged.select_dtypes("bool").columns
gdf_merged[bool_cols] = gdf_merged[bool_cols].astype(int)
gdf_merged
```

```python
gdf_merged.rename(columns = {
    "facgroup_CITY OPERATIONS" : "CITY OPERATIONS",
    "facgroup_EDUCATION" : "EDUCATION",
    "facgroup_HEALTH AND HUMAN SERVICES" : "HEALTH AND HUMAN SERVICES",
    "facgroup_OTHER" : "OTHER",
    "facgroup_PARKS AND RECREATION" : "PARKS AND RECREATION",
    "facgroup_PUBLIC SAFETY" : "PUBLIC SAFETY",
    "facgroup_TRANSPORTATION" : "TRANSPORTATION"
}, inplace=True)
```

## Aggregating data

```python
group_cols = ["CITY OPERATIONS", "EDUCATION", "HEALTH AND HUMAN SERVICES", 
              "OTHER", "PARKS AND RECREATION", "PUBLIC SAFETY", "TRANSPORTATION"]

info_cols = ["CD Name", "Shape_Leng", "Shape_Area", "city"]

agg_dict = {col: "sum" for col in group_cols} 
agg_dict["facname"] = "nunique"               
agg_dict.update({col: "first" for col in info_cols})  # pick first for these

gdf_agg = gdf_merged.groupby("DISTRICTCODE").agg(agg_dict).rename(
    columns={"facname": "total_facilities"}
).reset_index()
```

```python
gdf_agg['DISTRICTCODE'] = gdf_agg['DISTRICTCODE'].astype(int)
gdf_agg.head(5)
```

```python
gdf_final = pd.merge(gdf_agg, gdf[['BoroCD', 'geometry']], left_on='DISTRICTCODE', right_on='BoroCD', how='left')
gdf_final.head(5)
```

```python
gdf_final = gpd.GeoDataFrame(gdf_final, geometry="geometry", crs="EPSG:2263")
```

```python
gdf_final.to_file("facilities_w_districts.shp")
```
